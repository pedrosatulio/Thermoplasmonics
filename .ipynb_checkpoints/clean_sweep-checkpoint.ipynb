```python
import os
import math
import time
import numpy as npy
import scipy.signal as sp
import RefractiveIndex as ri
import matplotlib.pyplot as plt
import opticalCrossSection as ocs
import scipy.interpolate as interp
import numpy.polynomial.polynomial as poly
```

### Step 1: Parameter setting (manual)

```python
# Definição dos parâmetros de análise e importação dos parâmetros materiais

## Se a NP for uma nanoesfera, shell_material equivale ao material da esfera
shell_material = 'Au'                                    
core_material = 'Silica'
medium_material = 'Water'

## Se True, é nanocasca. Se False, é nanoesfera
isShell = True

## Incluir correção de drude?
drude = False

## Raio da da nanoesfera ou núcleo (nm)
a_min = 10
a_max = 95
a_res = 5

## Espessura da casca (nm)
t_min = 2
t_max = 32
t_res = 2

## Extensão da análise
lambda_min = 400                                         
lambda_max = 1100
lambda_ref = 1240
```

```python
path = "C:\\Users\\Ouvidoria\\Desktop\\Backup para prioritário\\Espectros Cascas\\"
```

### Step 2: Variable calculation (auto)

```python
#r = npy.linspace(a_min,a_max,int((a_max-a_min)/a_res+1))
th = npy.linspace(t_min,t_max,int((t_max-t_min)/t_res+1))
#lspr_d = npy.empty((len(r)))
#c_abs_d = npy.empty((len(r)))
#c_sca_s = npy.empty((len(r)))
#c_bck_s = npy.empty((len(r)))
#eta_abs_d = npy.empty((len(r)))
#eta_sca_s = npy.empty((len(r)))
#joule_d = npy.empty((len(r)))
#s2f_d = npy.empty((len(r)))
#FB = npy.empty((len(r)))
lspr_d = npy.empty((len(th)))
c_abs_d = npy.empty((len(th)))
c_sca_s = npy.empty((len(th)))
c_bck_s = npy.empty((len(th)))
eta_abs_d = npy.empty((len(th)))
eta_sca_s = npy.empty((len(th)))
joule_d = npy.empty((len(th)))
s2f_d = npy.empty((len(th)))
FB = npy.empty((len(th)))

a = 70
for xx in range(0,len(th)-1,1):
    t = th[xx]
    if isShell:
        b = a + t
    else:
        b = a
    wlt, N, N1t, N2t, mu1, mu2 = ri.setupRI(shell_material,core_material,medium_material,
                                            a,b,lambda_min,lambda_max,isShell,drude,0)
    wl = []
    N1 = []
    N2 = []
    for ii in range(0,len(wlt)-1,1):
        if (((1e9)*wlt[ii])>=lambda_min) and (((1e9)*wlt[ii])<=lambda_max):
            wl = npy.append(wl,wlt[ii])
            N1 = npy.append(N1,N1t[ii])
            N2 = npy.append(N2,N2t[ii])
    if isShell:
        Csca, Cext, Cabs, Cbck = ocs.shellCS(wl*(1e9),N1,N2,N,mu1,mu2,a,b)
    else:
        Csca, Cext, Cabs, Cbck = ocs.sphereCS(wl*(1e9),N2,N,mu2,a)
    wl = npy.delete(wl,len(wl)-1)
    Csca = npy.delete(Csca,len(Csca)-1)
    Cext = npy.delete(Cext,len(Cext)-1)
    Cabs = npy.delete(Cabs,len(Cabs)-1)
    fig1 = plt.figure(1)
    ax1 = plt.gca()
    lines1 = plt.plot((1e9)*wl,Csca,'#0072BD',(1e9)*wl,Cext,'#EDB120',(1e9)*wl,Cabs,'#D95319',markersize=4)
    plt.legend(lines1[:3], ['Scattering','Extinction','Absorption'], loc=0)
    #plt.legend(lines1[:3], ['Scattering','Extinction','Absorption'], bbox_to_anchor =(0.65, 1.25))
    yax = ax1.axes.get_yaxis()
    yax = yax.set_visible(False)
    plt.ylabel('Optical cross-sectional area (nm$^{2}$)')
    plt.xlabel('$\lambda$ (nm)')
    ax1.set_xlim([400,1100])
    ax1.set_title("r="+str(a)+"nm t="+str(t)+"nm", va='bottom')
    plt.savefig(path+'r='+str(a)+' t='+str(t)+'.png',dpi=300,format='png')
    plt.show()
    time.sleep(1)
    ##maxima = sp.find_peaks(Cext)
    #maxima = sp.find_peaks(Cabs)
    #dindex = maxima[0][len(maxima[0])-1]
    #lspr_d[xx] = (1e9)*float(wl[dindex])
    #c_abs_d[xx] = float(Cabs[dindex])
    #eta_abs_d[xx] = float(Cabs[dindex]/Cext[dindex])
    #vol = 4*math.pi*((b*(1e-9))**3)/3
    #Req = (3*vol/(4*math.pi))**(1/3)
    #joule_d[xx] = lambda_ref*(1e-9)*((1e-18)*float(Cabs[dindex]))/(2*math.pi*vol)
    #s2f_d[xx] = ((1e-18)*float(Cabs[dindex]))/Req
```
