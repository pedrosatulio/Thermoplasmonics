```python
import math
import numpy as npy
import scipy.signal as sp
import matplotlib.pyplot as plt
import scipy.interpolate as interp
import numpy.polynomial.polynomial as poly
```

```python
tabular = npy.loadtxt("Tabela_Cores.txt")
```

```python
L = tabular[:,0]
D = tabular[:,1]
LSPR = tabular[:,2]
```

```python
grid_x, grid_y = npy.mgrid[0:max(D):1024j, 0:max(L):1024j]
values = interp.griddata((D, L), LSPR, (grid_x, grid_y), method='cubic')
#values[npy.isnan(values)] = 520
values2 = npy.transpose(values)
```

```python
D_800 = []
L_800 = []
LSPR_800 = []
D_1064 = []
L_1064 = []
LSPR_1064 = []
```

```python
for m in range(0,len(grid_x)-1):
    for n in range(0,len(grid_y)-1):
        if (values2[m,n]>800-0.01 and values2[m,n]<800+0.01):
            D_800 = npy.append(D_800,grid_x[m,0])
            L_800 = npy.append(L_800,grid_y[0,n])
            LSPR_800 = npy.append(LSPR_800,values2[m,n])
        if (values2[m,n]>1064-0.03 and values2[m,n]<1064+0.03):
            D_1064 = npy.append(D_1064,grid_x[m,0])
            L_1064 = npy.append(L_1064,grid_y[0,n])
            LSPR_1064 = npy.append(LSPR_1064,values2[m,n])
```

```python
L_new_800 = npy.linspace(0, max(L_800), num=len(L_800)*50)
coefs_800 = poly.polyfit(L_800, D_800, 3)
ffit_800 = poly.polyval(L_new_800, coefs_800)
#plt.plot(L_new_800, ffit_800)
#plt.plot(L_800, D_800, marker='o', linestyle="None")
```

```python
L_new_1064 = npy.linspace(0, max(L_1064), num=len(L_1064)*50)
coefs_1064 = poly.polyfit(L_1064, D_1064, 3)
ffit_1064 = poly.polyval(L_new_1064, coefs_1064)
#plt.plot(L_new_1064, ffit_1064)
#plt.plot(L_1064, D_1064, marker='o', linestyle="None")
```

```python
fig1 = plt.figure(1)
ax1 = plt.gca()
ax1.set_xlim([8,62])
ax1.set_ylim([15,145])
plt.imshow(values2, origin='lower', extent=[0, max(D), 0, max(L)], aspect='auto', cmap=plt.get_cmap('jet'))
plt.plot(L_new_800, ffit_800, linewidth='3')
plt.plot(L_new_1064, ffit_1064, linewidth='3')
plt.legend(['  800 nm','1064 nm'], loc=4)
plt.ylabel('L (nm)')
plt.xlabel('D (nm)')
cbar1 = plt.colorbar()
cbar1.ax.set_ylabel('$\lambda_{LSPR}$ (nm)')
plt.savefig('AuNR_LSPR_Colormap.png', dpi=300)
plt.show()
```
